# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
df_train,df_test=train_test_split(wine3,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_train_scaled=pd.DataFrame(sc.fit_transform(df_train),columns=wine3.columns)
```

```python
df_test_scaled=pd.DataFrame(sc.transform(df_test),columns=wine3.columns)
```

```python
df_train_scaled.shape
```

```python
df_test_scaled.shape
```

```python
X_train=df_train_scaled.drop("pH",axis=1)
X_test=df_test_scaled.drop("pH",axis=1)
```

```python
y_train=df_train_scaled.pH
y_test=df_test_scaled.pH
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
rfe=RFE(lr,9)
```

```python
rfe=rfe.fit(X_train,y_train)
```

```python
rfe.support_
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#Select the columns which are true. Rfe.support shows both True and false. But when called, will take only True columns
selected_col=X_train.columns[rfe.support_]
selected_col
```

```python
# Convert the above selected_col list into dataframe
X_train_rfe=X_train[selected_col]
```

# Stats model

```python
import statsmodels.api as sm
```

```python
X_train_rfe_sm=sm.add_constant(X_train_rfe)
```

```python
linmod=sm.OLS(y_train,X_train_rfe_sm).fit()
```

```python
print(linmod.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_rfe.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe.values,i)for i in range(X_train_rfe.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by="VIF",ascending=False)
vif
```

```python
X_train_rfe_sm.drop("citric acid",axis=1,inplace=True)
```

```python
linmod1=sm.OLS(y_train,X_train_rfe_sm).fit()
```

```python
print(linmod1.summary())
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_rfe_sm.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe_sm.values,i)for i in range(X_train_rfe_sm.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by="VIF",ascending=False)
vif
```

```python
X_train_rfe_sm.head()
```

```python
y_train_sm_pred=linmod1.predict(X_train_rfe_sm)
```

```python
error_in_training=y_train-y_train_sm_pred
```

```python
sns.distplot(error_in_training)
```

# Accuracy for train data

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_sm_pred)
```

# Accuracy for test data

```python
X_test_sm=sm.add_constant(X_test)
```

```python
# X test will have all the columns initially. Now take the selected columns from X_train_rfe_sm
X_test_sm=X_test_sm[X_train_rfe_sm.columns]
```

```python
y_pred_test_sm=linmod1.predict(X_test_sm)
y_pred_train_sm=linmod1.predict(X_train_rfe_sm)
```
