VIF
Variance Inflation factor
It gives a basic representative idea about how much the feature variables are correlated with each other.
formula:
    VIF(i)=1/(1-R(i)**2)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
#Create a DataFrame taht will contain the names of all the feature variables and their respective VIFs
vif=pd.DataFrame()
vif['Features']=X_train.columns
vif['VIF']=[variance_inflation_factor(X_train.values,i)for i in range(X_train.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by="VIF",ascending=False)
vif
#Higher preference is given for p value as compared to VIF.
```
