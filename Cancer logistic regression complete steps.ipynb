## Clustering


```python
cancer_sc=sc.fit_transform(Cancer_DF)
cancer_sc
```

```python
cancer_cluster=pd.DataFrame(cancer_sc,columns=Cancer_DF.columns)
cancer_cluster.head()
```

```python
X=cancer_cluster
```

```python
from sklearn.cluster import KMeans
cluster_id=[]
K=list(range(1,5))
for i in K:
    km=KMeans(i).fit(X)
    y_km_pred=km.fit_predict(X)
    cluster_id.append(km.inertia_)
cluster_id
```

```python
plt.figure(figsize=(12,5))
plt.plot(K,cluster_id)
plt.show()
```
